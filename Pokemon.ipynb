# Classes 

```python
class Pokemon():
    '''
    Classe com o objetivo de servir de base para criação de todos os pokemons 
    '''
    def __init__(self, tipo, nome):
        '''
        Atributos Base de Todo Pokemon
        status: True, significa que o pokemon ainda pode batalhar; False, significa que 
        o pokemon está fora de combate 
        vida, são os pontos de vida do pokemon, caso chegue a zero o pokemon ficará fora de 
        combate 
        tipo, representa a qual tipo o pokemon faz parte 1 = grama, 2 = fogo, 3 = água
        nome, o nome que o pokemon possui 
        '''
        self.status = True
        self.vida = 100
        self.tipo = tipo
        self.nome = nome

class TipoFogo(Pokemon):
    '''
    classe que para criação de pokemons do tipo fogo
    '''
    def scratch(self, pokemon_inimigo):
        '''
        ataque de tipo normal
        '''
        dano = 20
        print(f'{self.nome} usa scratch!\n')

        pokemon_inimigo.vida -= dano

    def ember(self, pokemon_inimigo):
        '''
        ataque de tipo fogo 
        dano em tipo grama x2
        dano em tipo fogo x0.5
        dano em tipo água x0.5
        '''
        dano = 30
        print(f'{self.nome} usa ember!\n')

        if pokemon_inimigo.tipo == 1:
            pokemon_inimigo.vida -= dano * 2
            print('O ataque foi super efetivo\n')
        elif pokemon_inimigo.tipo == 2:
            pokemon_inimigo.vida -= dano * 0.5
            print('O ataque não foi muito efetivo\n')
        else:
            pokemon_inimigo.vida -= dano * 0.5
            print('O ataque não foi muito efetivo\n')

class TipoGrama(Pokemon):
    '''
    Classe para a criação de pokemons tipo planta
    '''
    def tackle(self, pokemon_inimigo):
        '''
        ataque do tipo normal 
        '''
        dano = 20

        print(f'{self.nome} tackle!\n')

        pokemon_inimigo.vida -= dano

    def leech_seed(self, pokemon_inimigo):
        '''
        ataque do tipo planta
        dano em tipo grama x0.5
        dano em tipo fogo x0.5
        dano em tipo água x2
        '''
        dano = 30

        print(f'{self.nome} usa leech seed!\n')

        if pokemon_inimigo.tipo == 3:
            pokemon_inimigo.vida -= dano * 2
            print('O ataque foi super efetivo\n')
        elif pokemon_inimigo.tipo == 2:
            pokemon_inimigo.vida -= dano * 0.5
            print('O ataque não foi muito efetivo\n')
        else:
            pokemon_inimigo.vida -= dano * 0.5
            print('O ataque não foi muito efetivo\n')

class TipoAgua(Pokemon):
    '''
    Classe para criação dos pokemons de tipo água 
    '''
    def tackle(self, pokemon_inimigo):
        '''
        ataque do tipo normal 
        '''
        dano = 20

        print(f'{self.nome} tackle!\n')

        pokemon_inimigo.vida -= dano

    def water_gun(self, pokemon_inimigo):
        '''
        ataque do tipo água
        dano em tipo grama x0.5
        dano em tipo fogo x2
        dano em tipo água x0.5
        '''
        dano = 30

        print(f'{self.nome} water gun!\n')

        if pokemon_inimigo.tipo == 2:
            pokemon_inimigo.vida -= dano * 2
            print('O ataque foi super efetivo\n')
        elif pokemon_inimigo.tipo == 1:
            pokemon_inimigo.vida -= dano * 0.5
            print('O ataque não foi muito efetivo\n')
        else:
            pokemon_inimigo.vida -= dano * 0.5
            print('O ataque não foi muito efetivo\n')
```

# Funções para o Jogo

```python
import random

def escolher_inicial():
    '''
    Função auxilia o player a escolher seu pokemon inicial
    Esta função não aceita argumentos, mas pede input para o usuário
    retorna sempre um pokemon 
    Exemplo:
        input: escolher_inicial() 
        output: objeto pokemon escolhido
    '''
    #pokemon que será escolhido pelo usário
    pokemon = None

    print(
        'Antes que comece sua aventura é necessário escolher seu primeiro pokemon!\n'
    )

    print(
        'Você possui três opções, use 1 para escolher Bulbasaur, 2 para Charmander e 3 para Squirtle\n'
    )

    #entrada da escolha do usário
    escolha_teste = input('Qual será sua escolha para parceiro? ')

    #validação da escolha do usuário
    if escolha_teste.isdigit:
        escolha = int(escolha_teste)
        #associando o pokemon com o input do usário
        if escolha == 1:
            pokemon = TipoGrama(1, 'Bubasaur')
        elif escolha == 2:
            pokemon = TipoFogo(2, 'Charmander')
        elif escolha == 3:
            pokemon = TipoAgua(3, 'Squirtle')
        else:
            print(
                'Por favor faça sua escolha, válida somente 1,2 e 3 são validos!\n'
            )
            escolher_inicial()
    else:
        print(
            'Por favor escolha seu inicial somente por meio de valores numéricos!\n'
        )

    print(f'Seu Pokemon escolhido foi {pokemon.nome}\n')
    return pokemon

def escolha_inimigo():
    '''
    Função que escolhe o inimigo do jogador(aleatório)
    Esta função não aceita argumentos e retorna uma string(Nome do inimigo) e seu pokemon
    Exemplo:
        input: escolha_inimigo()
        output: 'Erika', Bulbasaur
    '''
    #escolha aletória do inimigo
    codigo_inimigo = random.choice(range(2))
    #variaveis que serão retornadas
    pokemon_inimigo = None
    inimigo = None

    print(
        'Está na hora de testar sua habilidade como treinador pokemon, nada melhor que uma batalha para isso!\n'
    )

    #associação do código sorteado com as possibilidades de inimigos
    if codigo_inimigo == 0:
        inimigo = 'Erika'
        pokemon_inimigo = TipoGrama(1, 'Bubasaur')

        print("Seu inimigo será Erika, a princesa amante da natureza!\n")
        print(f'Seu o pokemon de seu oponente será {pokemon_inimigo.nome}\n')
    elif codigo_inimigo == 1:
        inimigo = 'Blaine'
        pokemon_inimigo = TipoFogo(2, 'Charmander')

        print("Seu inimigo será Blaine, o mestre de quiz cabeça quente!\n")
        print(f'Seu o pokemon de seu oponente será {pokemon_inimigo.nome}\n')
    else:
        inimigo = 'Misty'
        pokemon_inimigo = TipoAgua(3, 'Squirtle')

        print("Seu inimigo será Misty, a sereia moleca!\n")
        print(f'Seu o pokemon de seu oponente será {pokemon_inimigo.nome}\n')
    return inimigo, pokemon_inimigo

def inimigo_ataque(pokemon_jogador, inimigo, pokemon_inimigo):
    '''
    Função que determina o ataque do inimigo
    Esta função aceita como argumento o pokemon do jogador, o nome do inimigo e o pokemon do inimigo
    contudo não possui output, pois utiliza os métodos de ataque dos pokemons(inplace)
    Exemplo:
            input: inimigo_ataque(Charmander(objeto), Misty(inimigo), Squartle(objeto))
            output:
    '''
    #'Tomada de decisão' do inimigo baseado nos ataques disponiveis e o tipo de pokemon do usuário
    if inimigo == 'Erika':
        if pokemon_jogador.tipo == 3:
            pokemon_inimigo.leech_seed(pokemon_jogador)
        else:
            pokemon_inimigo.tackle(pokemon_jogador)
    elif inimigo == 'Misty':
        if pokemon_jogador.tipo == 2:
            pokemon_inimigo.water_gun(pokemon_jogador)
        else:
            pokemon_inimigo.tackle(pokemon_jogador)
    else:
        if pokemon_jogador == 1:
            pokemon_inimigo.ember(pokemon_jogador)
        else:
            pokemon_inimigo.scratch(pokemon_jogador)

def jogador_ataque(pokemon_jogador, pokemon_inimigo):
    '''
    Função pergunta ao jogador qual dos ataques deseja utilizar
    Esta função aceita como argumentos o pokemon do jogador e o pokemon do inimigo
    não possui nenhum retorno, pois se utiliza de métodos inplace 
    Exemplo:
            input: jogador_ataque(Charmander(objeto), Bubasaur(objeto))
            output:
    '''
    #mostra as opções de ataque do jogador baseado no pokemon do usuário
    if pokemon_jogador.tipo == 1:
        print('''Qual ataque deseja utiliza?\n
        1 para leech seed \n
        2 para tackle \n
        ''')
        #valida a opção escolhida pelo usuário
        opcao_teste = input()
        if opcao_teste.isdigit:
            opcao = int(opcao_teste)
            if opcao == 1:
                pokemon_jogador.leech_seed(pokemon_inimigo)
            else:
                pokemon_jogador.tackle(pokemon_inimigo)
        else:
            print('Por favor escolha uma opção númerica entre 1 e 2\n')
            jogador_ataque()
    elif pokemon_jogador.tipo == 2:
        print('''Qual ataque deseja utiliza?\n
        1 para ember\n
        2 para scratch\n
        ''')
        #valida a opção escolhida pelo usuário
        opcao_teste = input()
        if opcao_teste.isdigit:
            opcao = int(opcao_teste)
            if opcao == 1:
                pokemon_jogador.ember(pokemon_inimigo)
            else:
                pokemon_jogador.scratch(pokemon_inimigo)
        else:
            print('Por favor escolha uma opção númerica entre 1 e 2\n')
            jogador_ataque()
    else:
        print('''Qual ataque deseja utiliza?\n
        1 para water gun\n
        2 para tackle\n
        ''')
        #valida a opção escolhida pelo usuário
        opcao_teste = input()
        if opcao_teste.isdigit:
            opcao = int(opcao_teste)
            if opcao == 1:
                pokemon_jogador.water_gun(pokemon_inimigo)
            else:
                pokemon_jogador.tackle(pokemon_inimigo)
        else:
            print('Por favor escolha uma opção númerica entre 1 e 2\n')
            jogador_ataque()

def batalha(nome_jogador, pokemon_jogador, inimigo, pokemon_inimigo):
    '''
    Função que realiza a batalha entre jogador e inimigo
    Esta função usa como argumento o nome do jogador, o pokemon do jogador
    o nome do inimigo e seu pokemon; ela retorna True ou False, True: vitória do usuário
    False: vitória do inimigo
    Exemplo:
            input: batalha(Jogador, Charmander(objeto), Misty, Squartle(objeto))
            output: False
    '''

    print('A batalha irá começar!\n')

    #decisão se o jogador será o primeiro(True) ou não(False)
    vai_primeiro = random.choice([True, False])

    #reset dos status do pokemon do jogador, caso estaja jogando novamente
    pokemon_jogador.status = True
    pokemon_jogador.vida = 100

    #decisão do ciclo da batalha baseado na variável vai_primeiro
    if vai_primeiro:
        print(f'Você é o primeiro a atacar {nome_jogador}!\n')
        while pokemon_inimigo.status and pokemon_jogador.status:
            jogador_ataque(pokemon_jogador, pokemon_inimigo)
            inimigo_ataque(pokemon_jogador, inimigo, pokemon_inimigo)
            #finalização do ciclo de batalha caso a vida do pokemon zere
            if pokemon_jogador.vida <= 0:
                return False
            if pokemon_inimigo.vida <= 0:
                return True
    else:
        print(f'O primeiro a atacar é {inimigo}\n')
        while pokemon_inimigo.status and pokemon_jogador.status:
            inimigo_ataque(pokemon_jogador, inimigo, pokemon_inimigo)
            jogador_ataque(pokemon_jogador, pokemon_inimigo)
            #finalização do ciclo de batalha caso a vida do pokemon zere
            if pokemon_jogador.vida <= 0:
                return False
            if pokemon_inimigo.vida <= 0:
                return True

def jogar_novamente(nome_jogador):
    '''
    Função utlizada para encerrar ou continuar jogando
    Esta função aceita como argumento de entrada o nome do jogador
    e retorna True ou False, baseado na decisão do jogador 
    Exemplo:
            jogar_novamente(Jogador)
            output:True
    '''
    #Pergunta a decisão do jogador
    jogar_novamente = input('Deseja jogar novamente? S para Sim e N para Não ')
    #Baseado na decisão do jogador usa o retorno correto, True(continuar) e False(Parar)
    if jogar_novamente == 'S':
        return True

        print('Então vamos novamente!\n')
    elif jogar_novamente == 'N':
        return False

        print(f'Até mais ver {nome_jogador}\n')
    #validação da reposta do jogador para evitar erros
    else:
        print('Por favor responda somente S ou N, sem numeros ou espaços\n')
        #Chama a função novamente caso o jogaor erre a entrada no input
        jogar_novamente()
```

# Jogo

```python
#Pergunta do nome do usuário
nome_jogador = input('Olá treinador! Qual seu nome? ')

#variável para continuar/parar a batalha
continuar = True

print(f'Bem-vindo {nome_jogador}!\n')

#definição do pokemon do usuário
pokemon_jogador = escolher_inicial()

#ciclo de batalhas 
while continuar:
    #definição de inimigo e pokemon
    inimigo, pokemon_inimigo = escolha_inimigo()
    
    #resultado do embate 
    resultado = batalha(nome_jogador, pokemon_jogador, inimigo,
                        pokemon_inimigo)
    
    #print dos resultados da batalha
    if resultado:
        print(f'Parabéns pelo sua vitória {nome_jogador}!\n')
    else:
        print('Infelizmente hoje a vitória não foi sua\n')

    #mudança da variável de continuar/batalha    
    continuar = jogar_novamente(nome_jogador)
```
